```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import openrouteservice
import requests

np.random.seed(1000)

def get_uber_cost(emp_id, office_name,quarter,api_key):
    # read employee data from file
    emp_data = pd.read_csv('US_postal_codes_EID.csv')
    emp_info = emp_data[emp_data['Employee_ID'] == emp_id].iloc[0]
    #print(emp_info)
    emp_lat = emp_info['latitude']
    emp_lon = emp_info['longitude']

    # read office data from file
    office_data = pd.read_csv('US_salesforce_hubs-1.csv')
    office_info = office_data[office_data['Office Name'] == office_name].iloc[0]
    office_lat = office_info['Latitude']
    office_lon = office_info['Longitude']
    #print(office_info)

    # calculate distance and duration using TomTom API
    url = f'https://api.tomtom.com/routing/1/calculateRoute/{emp_lat},{emp_lon}:{office_lat},{office_lon}/json'
    params = {'key': api_key , "routeType": "fastest",
        "traffic": "true",
        "computeBestOrder": "false"}
    res = requests.get(url, params=params).json()
    distance_miles = round(res['routes'][0]['summary']['lengthInMeters'], 1) * 0.000621371
    distance_miles = round(distance_miles,2)
    distance = distance_miles
    duration_min = round(res['routes'][0]['summary']['travelTimeInSeconds'] * 0.0166667, 1)
    
    
    # calculate gas cost if distance is less than 220 miles
    print(f"The distance to the office: {distance_miles} miles")

    if distance_miles <= 220:
        
        print("Calculating ride share cost or the gas price, and the duration...")
        print(f"The Duration to the office: {duration_min} min")
        ride_share1 = round((0.90 + 1.90 + (duration_min * 0.09)+(distance_miles * 0.90)+3.50),2)
        
        # Load the gas price data
        gas_price_data = pd.read_excel('gas_prices.xlsx')
        # Get the employee's state
        employee_state = emp_data.loc[emp_data['Employee_ID'] == emp_id, 'state'].iloc[0]
        # Find the gas price in the employee's state
        gas_price = gas_price_data.loc[gas_price_data['state'] == employee_state, 'Regular'].iloc[0]
        # Calculate the cost
        cost1 = 25 * distance_miles / 100 * gas_price
        # Print the result
        #print('The gas price from employee {} to destination office {} is ${} if the employee decides to selfdrive.'.format(emp_id, office_name, round(cost1, 2)))
        #print(f"The cost if the Employee takes a Ride Share: ${ride_share1}") 
        cost = cost1
        ride_share1 = ride_share      
        return (ride_share,cost,distance)
    elif distance_miles > 220:
        # Load the airport data
        airport_data = pd.read_excel('Big_aiports_data.xlsx')
        # Find the nearest airport to the employee
        print('It is better to take the flight as the distance is too much. Does not make sense to drive to the office!')
        # find the nearest airport to the employee's location
        emp_location = (emp_lat, emp_lon)
        airport_data['Distance'] = airport_data.apply(lambda row: geodesic(emp_location, (row['Latitude'], row['Longitude'])).km, axis=1)
        nearest_airport = airport_data.loc[airport_data['Distance'].idxmin()]
        airport_name = nearest_airport['Origin']
        #airport_abbr = nearest_airport['Origin']
        airport_lat = nearest_airport['Latitude']
        airport_lon = nearest_airport['Longitude']
        
        # calculate distance and duration to the nearest airport
        airport_location = (airport_lat, airport_lon)
        #airport_address = Nominatim(user_agent='my_app').reverse(airport_location).address
        airport_distance_km = round(geodesic(emp_location, airport_location).km, 1)
        airport_distance_miles = round((airport_distance_km * 0.621371), 2)
        average_driving_speed_km = 90 # km/h
        average_driving_speed_mph = round(average_driving_speed_km * 0.621371, 2) # convert km/h to mph
        airport_duration_min = round((airport_distance_miles / average_driving_speed_mph) * 60, 1)
        Ride_Share1 = round((0.90 + 1.90 + (airport_duration_min * 0.09)+(airport_distance_miles * 0.90)+3.50),2)
        
        # Load the gas price data
        gas_price_data = pd.read_excel('gas_prices.xlsx')
        # Get the employee's state
        employee_state = emp_data.loc[emp_data['Employee_ID'] == emp_id, 'state'].iloc[0]
        # Find the gas price in the employee's state
        gas_price = gas_price_data.loc[gas_price_data['state'] == employee_state, 'Regular'].iloc[0]
        # Calculate the cost
        cost2 = 25 * airport_distance_miles / 100 * gas_price
        cost = cost2
        # read destination data from file
        destination_data = pd.read_csv('US_salesforce_hubs-1.csv')
        destination_info = destination_data[destination_data['Office Name'] == office_name].iloc[0]
        destination_lat = destination_info['Latitude']
        destination_lon = destination_info['Longitude']
        
        airport_data1 = pd.read_excel('Big_aiports_data.xlsx')
        airport_info1 = airport_data1[airport_data1['Origin'] == office_name].iloc[0]
        #print(airport_info1)
        airport_lat1 = airport_info1['Latitude']
        airport_lon1 = airport_info1['Longitude']
        airport_location1 = (airport_lat1,airport_lon1)

        # calculate distance from airport to destination
        destination_location = (destination_lat, destination_lon)
        destination_distance_km = round(geodesic(airport_location1, destination_location).km, 1)
        destination_distance_miles = round((destination_distance_km * 0.621371), 2)
        average_driving_speed_km = 90 # km/h
        average_driving_speed_mph = round(average_driving_speed_km * 0.621371, 2) # convert km/h to mph
        airport_duration_min1 = round((destination_distance_miles / average_driving_speed_mph) * 60, 1)
        Ride_Share2 = round((0.90 + 1.90 + (airport_duration_min1 * 0.09)+(destination_distance_miles * 0.90)+3.50),2)

        # Load the fare data
        fare_data = pd.read_csv('OG_file_OG_OG.csv')

        quarter = str(quarter) ###I want the quarter here to take in the same value I have passed in the def find_optimal_office(quarter, api_key)
        # Ask the user for the quarter
        #while True:
            #quarter = input("Enter the quarter (1, 2, 3, 4): ")
            #if quarter in ('1', '2', '3', '4'):
                #break
            #print("Invalid quarter. Please try again.")
        
        # Find the fare from employee airport to destination
        fare = fare_data.loc[(fare_data['Origin'] == airport_name ) & (fare_data['Dest'] == office_name) & (fare_data['Quarter'] == int(quarter)), 'MktFare'].values
        if len(fare) > 0 and pd.notna(fare[0]):
            fare = fare[0]
        else:
            if fare_data.loc[(fare_data['Origin'] == airport_name) & (fare_data['Dest'] == office_name),'MktFare'].isna().any():
                if int(quarter) == 1:
                    fare = np.random.randint(200, 601)
                elif int(quarter) == 2:
                    fare = np.random.randint(400, 751)
                elif int(quarter) == 3:
                    fare = np.random.randint(400, 601)
                elif int(quarter) == 4:
                    fare = np.random.randint(450, 751)
                
                #print(fare_data.loc[(fare_data['Origin'] == airport_name) & (fare_data['Dest'] == office_name)])
                #print('The fare employee, {} to destination {} is ${}'.format(airport_name, office_name, fare))
            else:
                fare = fare[0]

        
        # Print the fare
        #print(fare_data.loc[(fare_data['Origin'] == airport_name) & (fare_data['Dest'] == office_name)])
        ride_share = round(Ride_Share1 + Ride_Share2 + fare,2)
        distance = round(airport_distance_miles + destination_distance_miles,2)
        cost = round(cost2,2)
        
        # return results
        return (ride_share,cost,distance)
     
        
#calculate the ride share price from the airport to the destination office:
get_uber_cost(155, 'ATL',1, 'Ic1Y0VHsTzynY0lWKqMxG6p6LisiajJg')



def find_optimal_office(quarter, api_key):
    # read employee data from file
    emp_data = pd.read_csv('US_postal_codes_EID.csv')
    # select a random number of employees based on quarter number
    num_employees = np.random.randint(1, 10) 
    emp_ids = np.random.choice(emp_data['Employee_ID'], size=num_employees, replace=False)
    
    # read office data from file
    office_data = pd.read_csv('US_salesforce_hubs-1.csv')
    office_names = office_data['Office Name']
    
    # create dataframe to store results
    results = pd.DataFrame(columns=['Office Name', 'Ride Share', 'Cost', 'Distance'])
    
    # loop through each office and employee combination
    for office_name in office_names:
        ride_share_total = 0
        cost_total = 0
        distance_total = 0
        for emp_id in emp_ids:
            ride_share, cost, distance = get_uber_cost(emp_id, office_name,quarter, api_key)
            ride_share_total += ride_share
            cost_total += cost
            distance_total += distance
        num_emp = len(emp_ids)
        ride_share_weighted_avg = ride_share_total / num_emp
        cost_weighted_avg = cost_total / num_emp
        distance_avg = distance_total / num_emp
        score = 0.5*ride_share_weighted_avg + 0.3*cost_weighted_avg + 0.2*distance_avg
        results = results.append({'Office Name': office_name, 'Ride Share': ride_share_weighted_avg,
                               'Cost': cost_weighted_avg, 'Distance': distance_avg, 'Score': score},
                                 ignore_index=True)
    
    # return office with highest score
    return results.loc[results['Score'].idxmax(), 'Office Name']

find_optimal_office = find_optimal_office(1,'Ic1Y0VHsTzynY0lWKqMxG6p6LisiajJg')
print("The optimal office: {}",find_optimal_office)
```
